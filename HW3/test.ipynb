```python
import os
import xml.etree.ElementTree as ET

def convert_annotation(xml_path, output_path, image_path):
    # 指定以UTF-8編碼打開文件
    with open(xml_path, 'r', encoding='utf-8') as in_file:
        tree = ET.parse(in_file)
        root = tree.getroot()
        image_name = root.find('filename').text

        size = root.find('size')
        w = int(size.find('width').text)
        h = int(size.find('height').text)

        output_file = open(os.path.join(output_path, image_name.replace('.jpg', '.txt')), 'w')

        for obj in root.findall('object'):
            difficult = obj.find('difficult').text
            if difficult == '1':
                continue
            cls = obj.find('name').text
            if cls not in classes:
                continue
            cls_id = classes.index(cls)
            xmlbox = obj.find('bndbox')
            b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
            bb = convert((w, h), b)
            output_file.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')

        output_file.close()

def convert(size, box):
    dw = 1. / (size[0])
    dh = 1. / (size[1])
    x = (box[0] + box[2]) / 2.0 - 1
    y = (box[1] + box[3]) / 2.0 - 1
    w = box[2] - box[0]
    h = box[3] - box[1]
    x = x * dw
    w = w * dw
    y = y * dh
    h = h * dh
    return (x, y, w, h)

# 設置你的類別名稱列表
classes = ['container']  

# 設置XML文件所在目錄
xml_dir = 'yolov7-main\dataset\container\container-label\驗證集_xml'

# 設置輸出txt檔案的目錄 
output_dir = 'yolov7-main\dataset\container\container-label\驗證集'

# 設置圖片所在目錄
image_dir = 'yolov7-main\dataset\container\container-img\驗證集'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for xml_file in os.listdir(xml_dir):
    if xml_file.endswith(".xml"):
        xml_path = os.path.join(xml_dir, xml_file)
        convert_annotation(xml_path, output_dir, image_dir)
        print(f"Converted {xml_file} to YOLO format")
```

```python
import os

def merge_txt_files(input_dir, output_file):
    with open(output_file, 'w', encoding='utf-8') as outfile:
        for txt_file in sorted(os.listdir(input_dir)):
            if txt_file.endswith(".txt"):
                file_path = os.path.join(input_dir, txt_file)
                with open(file_path, 'r', encoding='utf-8') as infile:
                    outfile.write(infile.read() + "\n")
                print(f"Merged {txt_file}")

# 設置輸入TXT文件的目錄
input_dir = 'yolov7-main\dataset\container\container-label\驗證集'

# 設置輸出的總合併的TXT文件
output_file = 'yolov7-main\dataset\container\container-label\驗證集.txt'

merge_txt_files(input_dir, output_file)
print(f"All files have been merged into {output_file}")
```

```python
import os

def rename_images_in_directory(directory):
    # 獲取目錄中的所有文件，並按照文件名排序
    files = sorted(os.listdir(directory))
    
    # 初始化計數器
    counter = 2126
    
    # 遍歷文件
    for file in files:
        # 檢查文件是否為圖片文件（假設圖片文件為jpg格式）
        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
            # 構造新的文件名
            new_name = f"{counter}.jpg"
            # 獲取完整的舊文件路徑和新文件路徑
            old_file = os.path.join(directory, file)
            new_file = os.path.join(directory, new_name)
            # 重命名文件
            os.rename(old_file, new_file)
            print(f"Renamed {file} to {new_name}")
            # 增加計數器
            counter += 1

# 設置圖片所在目錄
image_dir = 'yolov7-main\dataset\container\container-img/val'

rename_images_in_directory(image_dir)
print("All images have been renamed.")
```

```python
import os

def generate_image_label_list(image_dir, output_file):
    # 获取目录中的所有文件，并按照文件名排序
    files = sorted(os.listdir(image_dir))
    
    # 打开输出文件
    with open(output_file, 'w', encoding='utf-8') as outfile:
        # 遍历文件
        for file in files:
            # 检查文件是否为图片文件（假设图片文件为jpg或png格式）
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                # 构造图片文件名和标签文件名
                image_file = os.path.join(image_dir, file)
                label_file = image_file.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt')
                # 检查对应的标签文件是否存在
                if os.path.exists(label_file):
                    # 将图片文件名和标签文件名写入输出文件
                    outfile.write(image_file + "\n")
                    outfile.write(label_file + "\n")
                else:
                    print(f"Warning: Label file {label_file} not found for image {image_file}")

# 设置图片和标签文件所在目录
image_dir = 'yolov7-main/dataset/container/container-img/train'

# 设置输出的文本文件
output_file = 'yolov7-main/dataset/container/container-img/train_list.txt'

# 生成图片和标签文件列表
generate_image_label_list(image_dir, output_file)
print(f"Generated {output_file}")
```

```python
import sys
sys.path.append('yolov7-main')

import cv2
import torch
from models.experimental import attempt_load
from utils.general import non_max_suppression, scale_coords
from utils.torch_utils import select_device

# 初始化
device = select_device('')  # 使用 GPU (如果可用)
weights = 'yolov7-main/runs/train\exp21\weights/best.pt'  # 權重文件路徑
model = attempt_load(weights, map_location=device)  # 載入模型
model.eval()

# 讀取 AVI 檔案
video_path = 'yolov7-main\影片資料集\video_0001.avi'  # AVI 檔案路徑
cap = cv2.VideoCapture(video_path)

# 取得影片的寬度和高度
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 設定輸出影片
output_path = 'output_video.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # 圖像預處理
    img = torch.from_numpy(frame).to(device)
    img = img.float() / 255.0
    if img.ndimension() == 3:
        img = img.unsqueeze(0)
    
    # 進行推論
    pred = model(img, augment=False)[0]
    pred = non_max_suppression(pred, 0.25, 0.45, classes=None, agnostic=False)

    # 處理結果
    for det in pred:
        if len(det):
            det[:, :4] = scale_coords(img.shape[2:], det[:, :4], frame.shape).round()
            for *xyxy, conf, cls in reversed(det):
                label = f'{cls} {conf:.2f}'
                cv2.rectangle(frame, (int(xyxy[0]), int(xyxy[1])), (int(xyxy[2]), int(xyxy[3])), (0, 255, 0), 2)
                cv2.putText(frame, label, (int(xyxy[0]), int(xyxy[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 將處理後的幀寫入輸出影片
    out.write(frame)

    # 顯示結果 (可選)
    cv2.imshow('Frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 釋放資源
cap.release()
out.release()
cv2.destroyAllWindows()
```
